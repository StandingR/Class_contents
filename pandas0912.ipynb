### apply(***) 

```python
df_kospi = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/KOSPI_stocks.csv')
df_kospi.head()
```

```python
#df_kospi['ChageRatio']
# 방법 1 함수 제작
def check_change(data):
 if data > 0:
  return '오른주식'
 else:    
    return '내린주식'
```

```python
# df_kospi['ChangeRatio'].apply(함수)
df_kospi['방법1'] = df_kospi['ChagesRatio'].apply(check_change) # apply 들어온 함수를 적용시키겠다.
df_kospi.head()
```

```python
# 방법 2, 람다는 조금 어렵다, 람다는 함수를 한줄로 쓰기 위한 식
df_kospi['방법2']= df_kospi['ChagesRatio'].apply(lambda x : '오른주식' if x > 0 else '내린주식')
df_kospi.head()
```

```python
import numpy as np
df_kospi['방법3']= np.where(df_kospi['ChagesRatio'] > 0, '오른주식', '내린주식')
df_kospi.head()
```

```python
(df_kospi['방법1'] == df_kospi['방법2']).value_counts()
```

```python
df_kospi['방법1'].value_counts()
```

```python
(df_kospi['방법2'] == df_kospi['방법3']).value_counts()
```

```python
"""con1 = df.query(f'EPS >= 0')

con1.query(f[EPS].rank(ascending=False) - [PER].rank(descending=True)) # apply 들어온 함수를 적용시키겠다.
con1.head()"""
```

```python
con1 = df.query(f'EPS >= 0')

con1['EPS_rank'] = con1['EPS'].rank(ascending=False)
con1['PER_rank'] = con1['PER'].rank(ascending=True)

# 순위 계산 결과를 이용하여 계산
con1['Rank_Difference'] = con1['EPS_rank'] - con1['PER_rank']

# 결과 확인
con1.head(10)
```

```python
df2 = df[df['EPS'] >= 0]
df2['EPS_rank'] = df2['EPS'].rank(ascending = False)
df2['PER_rank'] = df2['PER'].rank()
df2['PER_EPS_diff_rank'] = abs(df2['PER_rank'] - df2['EPS_rank'])
df2.sort_values('PER_EPS_diff_rank').head(10)
```

```python
market_price_change = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/market_price_change_2022.csv')
purchases_of_equities = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/purchases_of_equities_2022.csv')
```

```python
market_price_change.head(2)
```

```python
purchases_of_equities.head(2)
```

```python
merge1 = pd.merge(left = market_price_change, right =  purchases_of_equities, on = ['티커','종목명'])
merge1.head(2)
```

```python
merge1.shape # 기본으로 inner로 인식
```

```python
merge2 = pd.merge(left = market_price_change, right =  purchases_of_equities, on = ['티커','종목명'], how = 'left')
```

```python
merge2.shape # 공통된 두개 날라감
```

```python
merge3 = pd.merge(left = market_price_change, right =  purchases_of_equities, on = ['티커','종목명'], how = 'right')
merge3.shape
```

```python
merge4 = pd.merge(left = market_price_change, right =  purchases_of_equities, on = ['티커','종목명'], how = 'outer')
merge4.shape
```

## Pivot_table

테이터를 재구조화

```python
AAPL = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/AAPL.csv')
GOOGL = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/GOOGL.csv')
```

```python
GOOGL['Symbol'] = 'GOOGL'
AAPL['Symbol'] = 'APPL'
df = pd.concat([GOOGL, AAPL])
df
```

```python
pd.pivot_table(data = df, index = 'Date', columns = 'Symbol', values = 'Close')
```

```python
pd.pivot_table(data = df, index = ['Symbol','Date'], values = 'Close', aggfunc = 'min') # 각 값에 들어가는 ㄱㄱ밧이 명확할ㄱ떄
```

```python
df[['Date','Symbol','Close']].groupby(['Date','Symbol']).mean() # 일반적으로 group by 가 유리
```
